## 1b/c - Total capacity funding by core capacity (specified only)

```python
# Load data
fig1c = pd.read_csv('/Users/haileyrobertson/Documents/GitHub/ghs-tracking-analysis/analyses/Disbursed funds by core capacity by year/results/Disbursed funds by core capacity by year.csv')

# Clean data, converting year from numeric to datetime, and dropping the calculated columns
fig1c['Year'] = pd.to_datetime(fig1c['year'], format='%Y').dt.year
fig1c['Date'] = pd.to_datetime(fig1c['Year'].astype(str) + '-01-01')

fig1c = fig1c.melt(id_vars=['Year', 'Date', 'year'], var_name='Core capacity', value_name='Total disbursed')
fig1c = fig1c.fillna(0)

conditions = [
    fig1c['Core capacity'].str.startswith('P.'),
    fig1c['Core capacity'].str.startswith('D.'),
    fig1c['Core capacity'].str.startswith('R.'),
    fig1c['Core capacity'].str.startswith('CE'),
    fig1c['Core capacity'].str.startswith('PoE'),
    fig1c['Core capacity'].str.startswith('RE')
]
choices = ['Prevent', 'Detect', 'Respond', 'IHR Related Hazards','IHR Related Hazards','IHR Related Hazards']

# Use np.select to assign values based on conditions
fig1c['Theme'] = np.select(conditions, choices, default='NA')


fig1c = fig1c.sort_values(by="Year")
```

```python
fig1c.head()
```

```python

color_map = {
    'Prevent': '#3D4A8D',
    'Detect': '#A2BE2F',
    'Respond': '#CB494B',
    'IHR Related Hazards': '#633D71',
    '(?)':'#CACED3'}

# Create treemap
fig = px.treemap(fig1c,
                 path=['Theme', 'Core capacity'],
                 values='Total disbursed',
                 color="Theme"
                 color_discrete_map=color_map,
                 height=800,width=1300
                 )
year = fig1c['Year'].index.tolist()
fig.update_traces(tiling_packing='squarify')
fig.show()

fig.write_image("noyear_treemap.png") 
```

```python
color_map = {
    'Country': '#1A5784',
    'International': '#A2BE2F',
    'Other': '#CB494B',
    'Philanthropy': '#633D71'
}


# Create the Sankey diagram
sankey(
    left=fig3a["Funder"], fontsize=30, leftLabels=["Philanthropy","Other","International","Country"], right=fig3a["Recipient"],rightLabels=["Philanthropy","Other","International","Country"],leftWeight=fig3a["Total Value (2016-2019)"], rightWeight=fig3a["Total Value (2016-2019)"], colorDict=color_map
)


fig = plt.gcf()

# Set size in inches
fig.set_size_inches(16, 12)

# Set the color of the background to white
fig.set_facecolor("w")



# plt.show()

# plt.savefig('/Users/haileyrobertson/Documents/GitHub/ghs-tracking-analysis/figures/Fig3a.png', dpi=1000, facecolor='w', 
#             edgecolor='w', 
#             orientation='portrait')
```
